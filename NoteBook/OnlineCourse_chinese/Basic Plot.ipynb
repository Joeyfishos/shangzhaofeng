```python
t=np.arange(0.0,2.0,0.1)
s=np.sin(t*np.pi)
plt.plot(t,s,'r--',label='aaa')
plt.plot(t*2,s,'b--',label='bbb')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph')
plt.legend()
```
