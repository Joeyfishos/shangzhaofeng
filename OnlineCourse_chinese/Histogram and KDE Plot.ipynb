```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s=Series(np.random.randn(1000))
```

```python
plt.hist(s)
# 表示在-4到-3之间有1个，在-3到-2之间有2个，在-2到-2之间有18个，在-2到-1之间有62个。。的正态分布，
#频率分布直方图纵轴为频数，要有重复
```

```python
plt.hist(s,rwidth=0.9)
```

```python
s1= Series(np.arange(10))
```

```python
s1
```

```python
plt.hist(s1,rwidth=0.9)
#因为没u有重复
```

```python
plt.hist(s,rwidth=0.9,bins=20,color='r')#bins 默认是10个
```
