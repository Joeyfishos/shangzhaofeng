```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot():
    x=np.linspace(0,14,100)
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x,np.sin(x+i),label='sin(x)')
    plt.legend(bbox_to_anchor=(1.0,0.85),ncol=1)
```

```python
sinplot()
```
