# Data Preprocessing

```python
# Missing value
# becasuse from the info of the dataset, the attribute Age and Cabin has missing value,
# so using RFR meathod to predict missing Age value.
from sklearn.ensemble import RandomForestRegressor

def Set_Missing_Ages(df):
    age_df = df[['Age','Fare','Parch','SibSp','Pclass']]# extract all numeric attributes
    # seperate Age to two parts into matrix.
    known_age = age_df[age_df.Age.notnull()].as_matrix()
    unknown_age = age_df[age_df.Age.isnull()].as_matrix()
    
    # the target(Age)
    Y = known_age[:,0]
    # the feature atrribute
    X = known_age[:,1:]
    
    # fit X and Y into the RFR model
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
    rfr.fit(X,Y)
    
    # predict the age by using unknown feature attribute into model
    predicted = rfr.predict(unknown_age[:,1:])
    
    # fill the missing Age value with model value
    df.loc[ (df.Age.isnull()), 'Age' ] = predicted
    
    return df,rfr
    # only use loc meathod rather than df[df[]]
def Set_Cabin_Type(df):
    df.loc[ (df.Cabin.notnull()), 'Cabin' ] = "Yes"
    df.loc[ (df.Cabin.isnull()), 'Cabin' ] = "No"

    return df

data_train,rfr = Set_Missing_Ages(data_train)  #keep the rfr model for following test_data
data_train = Set_Cabin_Type(data_train)
```

```python
# factorization
# After finishing all missing data,in order to build Logistic regression model, model needed to
# delete all categoric value use dummy.(pd.get_dummies & pd.concat)
# categoric attributes are 'Cabin','Embarked','Sex','Pclass' , should be factorized

dummy_Cabin = pd.get_dummies(data_train['Cabin'],prefix='Cabin')
dummy_Embarked = pd.get_dummies(data_train['Embarked'],prefix='Embarked')
dummy_Sex = pd.get_dummies(data_train['Sex'], prefix= 'Sex')
dummy_Pclass = pd.get_dummies(data_train['Pclass'], prefix= 'Pclass')

# concatenate new dummyies and old dataset without useless attributes
# 'Pclass', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'
                                        
df = pd.concat([data_train, dummy_Cabin, dummy_Embarked, dummy_Sex, dummy_Pclass], axis = 1)
df.drop(['Pclass', 'Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True) 
# inplace=  change df inplace happended
df                                  
```
