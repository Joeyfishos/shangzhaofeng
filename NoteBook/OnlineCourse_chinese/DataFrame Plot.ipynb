```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= DataFrame(np.random.randint(1,10,40).reshape(10,4),columns=['A','B','C','D'])
```

```python
df
```

```python
df.plot()# 默认以columns来画图
```

```python
df.index
```

```python
for i in df.index:
    df.loc[i].plot(label=str(i))#用index横向画图
plt.legend(loc=3)
```

```python
df.plot(kind='bar')#同一个index下的不同columns的对比
```

```python
df.plot(kind='barh')# bar horizon 横向画图
```

```python
df.plot(kind='bar',stacked=True)# stacked 重叠吗？
```

```python
df.plot(kind='area')
```
