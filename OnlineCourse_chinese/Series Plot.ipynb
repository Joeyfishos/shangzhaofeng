```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s=Series(np.random.randn(1000))
s.plot(kind='kde',color='r')
```

```python
s1=Series([1,2,3,4,5])#累加
s1.cumsum()
```

```python
s1=Series(np.random.randn(10))
s2=Series(np.random.randn(10)).cumsum()
```

```python
s1
```

```python
s1.plot(kind='hist',color='r')
s2.plot(kind='line')
```

```python
fig,ax=plt.subplots(2,2)
s1.plot(ax=ax[0][0],color='r',label='s1',title='s1 title')
ax[0][0].legend()
plt.tight_layout()#将title展现出来
```

```python
figure,ax=plt.subplots(2,2)
ax[0][0].plot(s1,color='r',label='s1')#不支持title
ax[1][0].plot(s2)
```
