```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,5)#默认为50个
```

```python
x
```

```python
len(x)
```

```python
y1=np.sin(np.pi*x)
y2=np.sin(np.pi*x*2)
plt.plot(x,y1,'b--',label='sin(pi*x)')
plt.ylabel('y1 value')
plt.plot(x,y2,'r--',label='sin(pi*x*2)')
plt.ylabel('y2 value')
plt.xlabel('x value')
plt.title('this is x-y value')
plt.legend( loc='upper right')
```

```python
plt.show()
```

```python
plt.subplot(235)  #子图创建  两行三列的第五个
plt.plot(x,y1,'b--')
plt.subplot(232)
plt.plot(x,y2,'r--')
plt.subplot(233)
plt.plot(x,y1,'r*')
plt.subplot(234)
plt.plot(x,y2,'r*')
```
