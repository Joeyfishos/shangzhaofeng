```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
s= Series(np.random.randn(1000))
```

```python
s.plot(kind='kde')
```
